```
#librerias utilizadas

import pandas as pd
from google.colab import drive
import altair as alt
```

```
#nos conectamos a la BBDD de Google Drive y obtenemos los datos aleatorios generados

drive.mount('/gdrive')

random_data = pd.read_csv('/gdrive/My Drive/Pathologies Study/COVID-Study/data/random_data.csv', encoding = 'latin-1', sep = ";")
```

Vamos a generar un gráfico que nos permita visualizar cuantos elementos tenemos de cada cluster

Como se puede visualizar, tenemos 201 casos de Inflamación Intensa, 200 casos de Inflamación Leve y 200 casos de Inflamación Moderada, generando un dataset bastante compensado

```
alt.Chart(random_data).mark_bar().encode(
    y = 'Cluster',
    x = 'count(Cluster):Q'
)
```
